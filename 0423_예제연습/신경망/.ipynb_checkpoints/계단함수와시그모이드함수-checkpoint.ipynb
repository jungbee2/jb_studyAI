```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def step_function(x):
    return np.array(x>0, dtype=np.int32)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0,5.0,0.1)
y1 = step_function(x)
y2 = sigmoid(x)
plt.plot(x,y1)
plt.plot(x,y2,'--')
plt.ylim(-0.1,1.1)
plt.show()
```
